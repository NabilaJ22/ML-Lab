```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
```

```python
print(df.head())
```

# Calculate the mean of each feature grouped by species

```python
df_means = df.groupby('Species').mean().reset_index()
```

# a Bar plot for each feature

```python
df_melted = df_means.melt(id_vars='Species', var_name='Feature', value_name='Mean Value')
plt.figure(figsize=(12, 8))
sns.barplot(x='Feature', y='Mean Value', hue='Species', data=df_melted)
```

```python
plt.title('Average Feature Values by Species')
plt.xlabel('Features')
plt.ylabel('Mean Value')
```

```python
plt.show()
```
