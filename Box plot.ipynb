```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns=['Id', 'Species']))
```

```python
plt.title('Box Plot of Iris Dataset Features')
plt.xlabel('Features')
plt.ylabel('Value')
```

```python
plt.show()
```
